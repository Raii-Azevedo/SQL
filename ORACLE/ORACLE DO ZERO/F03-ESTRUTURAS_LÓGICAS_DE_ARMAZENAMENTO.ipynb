SEGMENTOS

São objetos dos bancos de dados Oracle

    -Tabelas - Indices - Views - Procedures
    
Todos são unidade Lógicas que ocupam espaço no HD (espaço físico)


ARQUIVOS FÍSICO x LÓGICOS

    - FISICOS:  Arquivos que possuem extensão .txt .docx .odp
    - LÓGICOS: Não possuem extensão definida.


COMO SEGMENTOS SÃO ARMAZENADOS? 

    - DataBlocks: Unidade mínime de armazenamento (Bloco de Dados)
    - Os blocos são compostos de (HEADER (Endereço do próximo bloco) | Espaço (vazio) redundância | Dados)


TAMANHOS PADRÃO

    - 2KB  8KB  4KB  16KB  32KB
  

BANCOS OLTP x OLAP

    - OTLT: Insert, Delete, Update
    - OLAP: Select
- Ou seja, OLAP é um bloco de leitura e o bloco OLTP é bloco de manipulação.
- O Tamanho 8KB é padrão, portanto híbrido. 2 e 4KB são direcionados para OLTP. E os tamanhos maiores são mais recomendados para blocos OLAP.


EXTENTS

O conjunto de DATABLOCKS forma uma EXTENT, ou seja, ocupa um tamanho (extenão).
- Quanto ao armazenamendo, os segmentos formam extensão, que não são necessáriamente contiguas, mas os blocos de dados que compõem uma extensão, são contíguos.
  

RESUMINDO

    - Datablock: Blocos de Disco
    - Segmentos: Objetos do banco de dados
    - Extents: Tamanho ocupado pelos segmentos
  - Ao criar uma tabela (Segment), ela ocupa Extent, que são compostos por datablock, ocupando espaço no disco.


  - Quando uma tabela é criada, aloca seu espaço em EXTENTS que são formados por DATABLOCKS contínuos. Quando essa tabela sofre um aumento de extensão, aloca mais extensão (EXTENT) que não necessáriamente está contínua ao EXTENT inicial.

  - Os EXTENTS são armazenados em TABLESPACES, que são criadas com um tamanho pré definifo que extensts lives. A medida que os objetos vão sendo alocados, esses extents vão sendo alocados.
